Realicé el ejercicio empleando la API: Free Fake REST API for Placeholder JSON Data (https://dummyjson.com)

1. Importación de librerias

```python
import requests
import pandas as pd
```

2. Empleando el endpoint /products se consultaron los productos disponibles en la API. Con el parámetro limit=100 se obtuvieron los primeros 100 registros. Posteriormente, se realizó la primera consulta a la API y se validó el status_code para confirmar que la respuesta fue exitosa.

```python
base_url = "https://dummyjson.com"
endpoint = "/products"
url = base_url + endpoint+"?limit=100"
response = requests.get(url)
print("Status code:", response.status_code)
```

3. La respuesta en formato JSON se convirtió en un DataFrame para poder visualizar la información de manera más organizada. Sin embargo, se observó que los datos se encontraban dentro de un diccionario, por lo que se extrajo la información correspondiente a los productos para poder trabajar con ella y utilizarla en el análisis.

```python
data = response.json()
df = pd.DataFrame(data)
df.head()
```

```python
df = pd.DataFrame(data["products"])
df.head()
```

4. Se validó la información del DataFrame, identificando los tipos de datos de cada columna, la presencia de valores vacíos, así como la cantidad y el tipo de columnas disponibles.

```python
df.info()
```

5. Se consultaron desde la API las categorías disponibles de los productos, utilizando el endpoint:  */products/categories*

```python
endpoint_categorias = "/products/categories"
url_categorias = base_url + endpoint_categorias
reponse_categorias = requests.get(url_categorias)
categorias = reponse_categorias.json()
print(categorias)
```

6. Se eliminaron las columnas: description, sku, weight, dimensions, arrantyInformation, shippingInformation, returnPolicy, reviews, meta, images y thumbnail, debido a que la informacion que contienen no resulta relevante para el analisis.

```python
df = df.drop(columns=["description","sku","weight","dimensions","warrantyInformation","shippingInformation","returnPolicy","reviews","meta","images","thumbnail"])
df
```

7. En el paso 4 se evidenció que la columna brand no se encontraba completamente diligenciada, ya que presentaba valores nulos. Por esta razón, se procedió a reemplazarlos con el valor "Unknown", con el fin de mantener la integridad del dataset y permitir el análisis completo de los datos sin pérdida de registros.

```python
df["brand"] = df["brand"].fillna("Unknown")
df.head()
```

8. Se creo la columna indice para combinar la calificación del producto y la disponibilidad en inventario, lo anterior con el objetivo de identificar productos que no solo son bien valorados por los usuarios, sino que también tienen suficiente stock para su venta.

```python
df["indice"] = df["rating"] * df["stock"]
df
```

9. Se creó la columna relación precio vs rating para analizar si un producto es costoso en comparación con la valoración que recibe, lo que permite identificar aquellos productos que pueden percibirse como “caros” o que tienen una menor relación calidad-precio.

```python
df["relacion_precio_rating"] = df["price"] / df["rating"]
df
```

10. Se cambiaron los nombres de las columnas con el objetivo de facilitar la comprensión de los datos y mejorar su uso en el análisis dentro de Power BI, haciendo el dataset más organizado y fácil de interpretar.

```python
df = df.rename(columns={
    "id": "ID",
    "title": "Producto",
    "category": "Categoria",
    "price": "Precio unitario",
    "discountPercentage": "Descuento",
    "rating": "Puntaje",
    "stock": "Cantidad disponible",
    "tags": "Etiquetas",
    "brand": "Marca",
    "availabilityStatus": "Disponibilidad",
    "minimumOrderQuantity": "Cantidad minima",
    "indice": "Indice Puntaje vs Stock",
    "relacion_precio_rating": "Relacion Precio vs Puntaje",
})
df
```

```python
df["Precio unitario"] = df["Precio unitario"].round(2)
df["Relacion Precio vs Puntaje"] = df["Relacion Precio vs Puntaje"].round(2)
df
```
