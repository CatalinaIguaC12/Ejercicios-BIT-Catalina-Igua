**CONTAMINACIÓN DEL AIRE**

La contaminación del aire es una de las mayores amenazas ambientales para la salud. La exposición a contaminantes como el material particulado (PM2.5),el dióxido de nitrógeno (NO₂), el ozono (O₃) y el monoxido de carbono (CO) puede causar:

1. Afectación a pulmones y corazón, pueden causar enfermedades graves y muerte prematura.
2. Irrita ojos y vías respiratorias, dificultan la respiración.
3. Inflama los pulmones y empeora el asma, aumentan infecciones respiratorias.
4. Reducen el oxígeno en la sangre; pueden causar mareos, desmayos y en casos graves, muerte.

El AQI(Air Quality Index) es un valor numérico adimensional que resume la concentración de varios contaminantes atmosféricos en un solo indicador fácil de interpretar.

Se estima mediante funciones estandarizadas definidas por agencias ambientales como la Environmental Protection Agency (EPA) y la Organización Mundial de la Salud (OMS).

Los rangos del AQI son:

1. 0- 50: Aire limpio sin riesgo para la salud
2. 51 – 100: Riesgo bajo para la población general.
3. 101 – 150: Puede afectar a personas vulnerables.
4. 151 – 200: Afecta a toda la población.
5. 201 – 300: Riesgo significativo para la salud.
6. 301 – 500: Emergencia sanitaria.

Se eliminaron los registros sin pais y sin ciudad, por que sin esta informacion no es posible realizar comparaciones.

```python
datos=datos.dropna(subset=['City'])
datos.isnull().sum()
```

Se utilizó la función .duplicated() para verificar si existían ciudades repetidas en el conjunto de datos, con el fin de determinar la necesidad de calcular promedios. Sin embargo, no se identificaron duplicados en esta información.

```python
datos['City'].duplicated().sum()
```

Para evitar confusiones en el análisis de la información, se empleó la función .replace() con el fin de modificar los valores de la columna AQI Category y así facilitar su interpretación.

Para ello fue necesario:

1. Identificar las diferentes categorias que tiene la columna AQI Category, empleando .value_counts().

```python
datos['AQI Category'].value_counts()
```

2. Se empleó la función .replace() para cambiar las catergorias de la columna AQI Category

```python
datos['AQI Category'] = datos['AQI Category'].str.strip()
```

```python
datos['AQI Category']= datos['AQI Category'].str.replace('Unhealthy for Sensitive Groups','Dañino para grupos sensibles')
datos['AQI Category']= datos['AQI Category'].str.replace('Dañino for Sensitive Groups','Dañino para grupos sensibles')
datos['AQI Category']= datos['AQI Category'].str.replace('Very Unhealthy','Muy dañino')
datos['AQI Category']= datos['AQI Category'].str.replace('Very Dañino','Muy dañino')
datos['AQI Category']= datos['AQI Category'].str.replace('Unhealthy','Dañino')
datos['AQI Category']= datos['AQI Category'].str.replace('Good','Bueno')
datos['AQI Category']= datos['AQI Category'].str.replace('Moderate','Moderado')
datos['AQI Category']= datos['AQI Category'].str.replace('Hazardous','Peligroso')
datos['AQI Category'].value_counts()
```

Se incorporaron dos nuevas columnas en la tabla, Contaminante_Mayor y Valor_Max_Contaminante, como un resumen de los datos más relevantes por cada ciudad, con el fin de facilitar el análisis de la información.

```python
datos['Contaminante_Mayor'] = datos[
    ['CO AQI Value', 'Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value']
].idxmax(axis=1)

datos
```

```python
datos['Valor_Max_Contaminante'] = datos[
    ['CO AQI Value', 'Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value']
].max(axis=1)

datos
```
